## Import the dataset

```python
import pandas as pd

file_path = "/content/drive/MyDrive/Experience-Salary.csv"

df = pd.read_csv(file_path)

df.head()
```

## Clean the Dataset

```python
df = df.round(0)  # round all the decimals values to 0 decimal

df.head()
```

## Spliting the dataset into the x and y axis

```python
x = df[['exp(in months)']]
y = df['salary(in thousands)']
```

```python
x.head()
```

```python
y.head()
```

## Split the dataset into the training and the testing dataset

```python
# import the sci-kit learn

from sklearn.model_selection import train_test_split

x_train, x_test , y_train , y_test = train_test_split(x,y,test_size=0.2, random_state=0)
```

## Fit the Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression().fit(x_train,y_train)
model
```

## Plot the training data

```python
import matplotlib.pyplot as plt

plt.scatter(x_train,y_train,color="red")

# view the line plot by the linear regression model

plt.plot(x_train, model.predict(x_train),color='green')

plt.xlabel("Experience (in months)")
plt.ylabel("Salary (in thousands)")
plt.title("Salary vs Experience (Training Dataset)")

plt.show()
```

### Plot the Testing Dataset

```python
plt.scatter(x_test, y_test, color='blue')
# view the plot line in the testing dataset
plt.plot(x_test, model.predict(x_test),color='red')
plt.xlabel("Experience (in months)")
plt.ylabel("Salary (in thousands)")
plt.title("Salary vs Experience (Testing Dataset)")

plt.show()
```

### Evaluate the model fitness (calcuting the regression score)

```python
print("Training Score:", model.score(x_train, y_train))
print("Testing Score:", model.score(x_test, y_test))
```

### Evaluate the model on the unknown values

```python
model.predict([[10]])   # salary with experience of 10 years
```
